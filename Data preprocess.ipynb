```python
dataset.drop(columns=['id'], inplace=True)
```

```python
for col in ['pcv', 'wc', 'rc']:
    dataset[col] = pd.to_numeric(dataset[col], errors='coerce')
```

```python
Qual = dataset.select_dtypes(include='object').columns
Quan = dataset.select_dtypes(include=['float64', 'int64']).columns
```

```python
Quan
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp_freq = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imp_freq.fit(dataset[Qual])
df_Qual=imp_freq.transform(dataset[Qual])
df_Qual=pd.DataFrame(df_Qual,columns=Qual)
dataset[Qual]=df_Qual
```
